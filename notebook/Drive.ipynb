```python
from dotenv import load_dotenv
import os
load_dotenv()
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = ""
FOLDER_ID = os.getenv("FOLDER_ID")
```

```python
from langchain_google_community import GoogleDriveLoader
from langchain_community.document_loaders import UnstructuredFileIOLoader

print("Initializing Google Drive loader...")
loader = GoogleDriveLoader(
    folder_id=FOLDER_ID,
    file_types=["document", "pdf"],  # Optional: specify file types to load. Defaults to all types.
    
    file_loader_cls=UnstructuredFileIOLoader,
    token_path="D:/Gen AI/RAG/Simple-RAG/google_token.json",
    credentials_path="D:/Gen AI/RAG/credentials.json",
    
    recursive=False,
)
# docs = loader.load()
# print(f"Loaded {len(docs)} document(s) from Google Drive folder")

# Replace docs = loader.load() with this:
print("Starting load...")
docs = []
for doc in loader.lazy_load():
    docs.append(doc)
    print(f"Loaded: {doc.metadata.get('title', 'Unknown File')}")

print(f"Finished! Total docs: {len(docs)}")
```
