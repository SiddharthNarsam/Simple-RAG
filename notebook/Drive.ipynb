```python
from dotenv import load_dotenv
import os
load_dotenv()
FOLDER_ID = os.getenv("FOLDER_ID")
```

```python
print("Initializing Google Drive loader...")
from langchain_google_community import GoogleDriveLoader
loader = GoogleDriveLoader(
    folder_id=FOLDER_ID,
    file_types=["document", "pdf"],
    token_path="D:\\Gen AI\\RAG\\Simple-RAG\\google_token.json",
    credentials_path="..../credentials.json"
    # Optional: configure whether to recursively fetch files from subfolders. Defaults to False.
    recursive=False,
)
# docs = loader.load()
# print(f"Loaded {len(docs)} document(s) from Google Drive folder")

# Replace docs = loader.load() with this:
print("Starting load...")
docs = []
for doc in loader.lazy_load():
    docs.append(doc)
    print(f"Loaded: {doc.metadata.get('title', 'Unknown File')}")

print(f"Finished! Total docs: {len(docs)}")
```
